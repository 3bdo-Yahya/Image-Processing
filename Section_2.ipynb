```python
import cv2 # OpenCV library for image processing
import matplotlib.pyplot as plt # For displaying images
```

```python
from skimage import data # to import sample images
image = data.astronaut()   # ready sample (RGB numpy array)
plt.imshow(image); plt.axis('off')
```

```python
image.shape # Get the dimensions of the image
```

```python
# image resizing function:

# cv2.resize (src, dsize, fx, fy, interpolation)
# dsize = (width, height) but target image shape will be (height, width, 3)
# If dsize is None, it is computed from fx and fy
# INTER_AREA is good for shrinking
# INTER_CUBIC or INTER_LINEAR is good for zooming
```

```python
# resize image to 10% of original size

shrinked_image = cv2.resize(image, None, fx=0.1, fy=0.1, interpolation=cv2.INTER_AREA)
```

```python
# resize image to 200% of original size

zoomed_image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
# or
# zoomed_image = cv2.resize(image, (image.shape[1]*2, image.shape[0]*2), interpolation=cv2.INTER_CUBIC)

# INTER_CUBIC is slower but better quality

# using linear:
linear_200 = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
```

```python
# inter linear to 50% size:
linear_image = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR) # Default
# inter nearest to 50% size:
nearest_image = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)

# diff btw linear and nearest is that linear uses weighted avg of 4 nearest pixels
# while nearest just picks the nearest pixel value
```

```python
Titles = ['Original Image', 'Linear (50%)', 'Nearest (50%)', 'Shrinked (10%)', 'Linear (200%)', 'Zoomed (200%)']
Images = [image, linear_image, nearest_image, shrinked_image, linear_200, zoomed_image]
```

```python
for i in range(len(Images)):
    plt.subplot(2, 3, i+1)
    plt.imshow(Images[i])
    plt.title(Titles[i])
    plt.axis('off')
plt.tight_layout()
```
