```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage import data # to import sample images
image = data.cat()   # ready sample (RGB numpy array)
plt.imshow(image); plt.axis('off')

# if you want to use your own image, uncomment the following line:
# image = cv2.imread('path_to_your_image.jpg')  # read image in ( BGR ) format
```

```python
image.shape  # (height, width, channels)
```

```python
image_GBR = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)  # convert RGB to BGR
plt.imshow(image_GBR); plt.axis('off')
```

```python
# if you read an image using OpenCV, it will be in BGR format by default
# To convert it to RGB format, you can use the following line:

image_RGB = cv2.cvtColor(image_GBR, cv2.COLOR_BGR2RGB)  # convert BGR to RGB
plt.imshow(image_RGB); plt.axis('off')
```

```python
image_Gray = cv2.cvtColor(image_GBR, cv2.COLOR_BGR2GRAY)  # convert BGR to Grayscale

# Displaying without colormap (cmap) will show it in a default colormap (viridis)
plt.imshow(image_Gray); plt.axis('off')
```

```python
image_Gray = cv2.cvtColor(image_GBR, cv2.COLOR_BGR2GRAY)  # convert BGR to Grayscale
plt.imshow(image_Gray, cmap='gray'); plt.axis('off')

# Why we use cmap='gray'?
# Correct display:

# For single-channel (grayscale) images:
# cmap='gray' tells imshow to map intensity  shades of gray.
# Without it, imshow uses the default colormap (viridis)
# and your grayscale image may display in false color.
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_RGB); plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(image_GBR); plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(image_Gray, cmap='gray'); plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
B, G, R = cv2.split(image_GBR)  # split into B, G, R channels

print(image_GBR.shape)  # (height, width, channels)
print(B.shape)  # (height, width)
print(G.shape)  # (height, width)
print(R.shape)  # (height, width)
```

```python
print(image_GBR)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(R, cmap='Reds'); plt.title('Red Channel'); plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(B, cmap='Blues'); plt.title('Blue Channel'); plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(G, cmap='Greens'); plt.title('Green Channel'); plt.axis('off')
```

```python
image_1 = cv2.blur(image_GBR, (15, 15))  # apply Gaussian blur with a 15x15 kernel
plt.imshow(image_1); plt.axis('off')
```

```python
image_2 = cv2.GaussianBlur(image_GBR, (15, 15), 0)  # apply Gaussian blur with a 15x15 kernel
plt.imshow(image_2); plt.axis('off')
```

```python
image_3 = cv2.medianBlur(image_GBR, 7)  # apply Median blur with a kernel size of 7
plt.imshow(image_3); plt.axis('off')
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(image_1); plt.title('Averaging Blur'); plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(image_2); plt.title('Gaussian Blur'); plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(image_3); plt.title('Median Blur'); plt.axis('off')

plt.tight_layout(); plt.show()
```
